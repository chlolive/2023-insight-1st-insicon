# 유의미한 칼럼 추출하기

```python
data_1= data_1.drop(columns='LoanID')
data_1.head()
```

LoanID는 대출자를 구별하기 위한 식별 번호로, 채무불이행에 대한 설명변수로 적합하지 않기에 열 삭제

```python
numeric_data = data_1.select_dtypes(include=[np.number])  # 숫자형 데이터만 선택
correlation_matrix = numeric_data.corr()  # 숫자형 데이터의 상관관계 행렬 계산

plt.figure(figsize=(18, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlGn', linewidths=0.2)
plt.show()
```

heatmap으로 살펴본 결과, 수치형 데이터 항목 중에서는 상관계수가 유의미하게 높은 독립변수는 없는 것으로 확인된다.

새로 발견한 점
DTIRatio가 낮은 그룹일수록 Debt의 최대치도 적고 Debt의 범위가 아래쪽으로 몰려있다. -> 즉, DTIRatio가 낮은 그룹은 빚도 적다.  
Debt의 양이 절대적으로 많을수록 DTI ratio도 안 좋을 확률이 높다.  
그런데 한 가지 이상한 점이 '왜 각 그룹에서 Debt이 적을 수록 Default가 1일 확률이 높아질까?'  
상식적으로 debt이 많을수록 상환을 못할 확률이 높은데, 이에 대해 더 알아보기 위해 아래와 같이 debt를 구간 별로 나누기로 함.
